<h3>Étape 2: Analyse du contenu HTML de page de base de chaque catégorie

```python
import requests
from bs4 import BeautifulSoup
all_products=[] # la liste de stockage de produits
ids_reference =[] #la liste a générer pour savoir remplir le fichier de notes des utilisateurs
#liste de urls de page de base de chaque catégorie
URL_base_list =["https://www.jumia.com.tn/epicerie/",
                "https://www.jumia.com.tn/maison-cuisine-jardin/",
                "https://www.jumia.com.tn/beaute-hygiene-sante/",
                "https://www.jumia.com.tn/telephone-tablette/",
                "https://www.jumia.com.tn/fashion-mode/",
                "https://www.jumia.com.tn/ordinateurs-accessoires-informatique/",
                "https://www.jumia.com.tn/electronique/",
                "https://www.jumia.com.tn/jeux-videos-consoles/",
                "https://www.jumia.com.tn/sports-loisirs/",
                "https://www.jumia.com.tn/automobile-outils/",
                "https://www.jumia.com.tn/terrasse-jardin-exterieur/",
                "https://www.jumia.com.tn/bebe-puericulture/",
                "https://www.jumia.com.tn/animalerie/",
                "https://www.jumia.com.tn/industriel-scientifique/",
                "https://www.jumia.com.tn/instruments-musique/",
                "https://www.jumia.com.tn/divers/",
                "https://www.jumia.com.tn/livres-papeterie/",
                "https://www.jumia.com.tn/jeux-et-jouets/"]
#extraction de donées de chaque page html.
for url in URL_base_list:
    r = requests.get(url)
    #print(r.content)
    soup = BeautifulSoup(r.content, 'html5lib')
    #print(soup.prettify()) # afficher le squelette de l'objet soupe.
    table = soup.find('div', attrs = {'class':'-paxs row _no-g _4cl-3cm-shs'})
    for row in table.findAll('article',attrs = {'class':'prd _fb col c-prd'}):
        product = {}
        reference ={}
        product['product_id'] = row.a['data-id']
        reference['product_id'] = row.a['data-id']
        product['product_name'] = row.a['data-name']
        product['product_category'] = row.a['data-category']
        product['product_brand'] = row.a['data-brand']
        sub_table=row.find('div', attrs = {'class' : 'prc'})
        product['product_price'] = sub_table.text
        avis =row.find('div', attrs={'class':'rev'})
        if(avis):
            avis_string=avis.text.split()[3]
        else:
            avis_string = '0000'
        product['reviews_number'] = avis_string[2:(len(avis_string)-1)]
        reference['reviews_number'] = avis_string[2:(len(avis_string)-1)]
        all_products.append(product)
        ids_reference.append(reference)
    
```

```python
from bs4 import BeautifulSoup
# creation d'une liste pour automatiser la réponse de site web
index_list=[str(i) for i in range(2,51)]
URL_pages_list =["https://www.jumia.com.tn/epicerie/?page=",        
                 "https://www.jumia.com.tn/maison-cuisine-jardin/?page=",
                 "https://www.jumia.com.tn/beaute-hygiene-sante/?page=",
                 "https://www.jumia.com.tn/telephone-tablette/?page=",
                 "https://www.jumia.com.tn/fashion-mode/?page=",
                 "https://www.jumia.com.tn/ordinateurs-accessoires-informatique/?page=",
                 "https://www.jumia.com.tn/electronique/?page=",
                 "https://www.jumia.com.tn/jeux-videos-consoles/?page=",
                 "https://www.jumia.com.tn/sports-loisirs/?page=",
                 "https://www.jumia.com.tn/automobile-outils/?page=",
                 "https://www.jumia.com.tn/terrasse-jardin-exterieur/?page=",
                 "https://www.jumia.com.tn/bebe-puericulture/?page=",
                 "https://www.jumia.com.tn/jeux-et-jouets/?page=",
                 "https://www.jumia.com.tn/divers/?page=",
                 "https://www.jumia.com.tn/livres-papeterie/?page=",
                 "https://www.jumia.com.tn/instruments-musique/?page=",
                 "https://www.jumia.com.tn/industriel-scientifique/?page="]
for url in URL_pages_list:   
    for j in index_list:
        print(j)
        URL =url+j+"#catalog-listing"
        print("scraping "+URL)
        r = requests.get(URL)
        soup = BeautifulSoup(r.content, 'html5lib')
        #print(soup.prettify()) # afficher le squelette de l'objet soupe.
        table = soup.find('div', attrs = {'class':'-paxs row _no-g _4cl-3cm-shs'})
        if (table):
            for row in table.findAll('article',attrs = {'class':'prd _fb col c-prd'}):
                product = {}
                reference ={}
                product['product_id'] = row.a['data-id']
                reference['product_id'] = row.a['data-id']
                product['product_name'] = row.a['data-name']
                product['product_category'] = row.a['data-category']
                product['product_brand'] = row.a['data-brand']
                sub_table=row.find('div', attrs = {'class' : 'prc'})
                product['product_price'] = sub_table.text
                avis =row.find('div', attrs={'class':'rev'})
                if(avis):
                    avis_string=avis.text.split()[3]
                else:
                    avis_string = '0000'
                product['reviews_number'] = avis_string[2:(len(avis_string)-1)]
                reference['reviews_number'] = avis_string[2:(len(avis_string)-1)]
                all_products.append(product)
                ids_reference.append(reference)
                
```

<h3>Étape 4: Enregistrez toutes les données dans un fichier CSV

```python
import csv
filename = 'all_products.csv'
with open(filename, 'w', newline='', encoding="utf-8") as f:
    w = csv.DictWriter(f,['product_id','product_name','product_category','product_brand', 'product_price', 'reviews_number'])
    w.writeheader()
    for product in all_products:
        w.writerow(product)
```

```python
filename = 'ids_reference.csv'
with open(filename, 'w', newline='', encoding="utf-8") as f:
    w = csv.DictWriter(f,['product_id','reviews_number'])
    w.writeheader()
    for product in ids_reference:
        w.writerow(product)
```
